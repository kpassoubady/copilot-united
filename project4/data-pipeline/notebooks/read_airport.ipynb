```python
import pandas as pd

df = pd.read_csv('data/airports.csv')
df.head()
```

```python
# I want to see the tail of the dataframe. How can I do that?
df.tail()
```

```python
# I want to filter the dataframe to only include airports in the state CA. How can I do that?
df[df['state'] == 'CA']
```

```python
# i want to filter the dataframe to only include airports in the state CA and city Los Angeles. How can I do that?
df[(df['state'] == 'CA') & (df['city'] == 'Los Angeles')]
```

```python
# Print the number of airports in the state CA
print(len(df[df['state'] == 'CA']))
```

```python
# Find the duplicate rows in the dataframe
df[df.duplicated()]
```

```python
# drop the duplicate rows
df.drop_duplicates(inplace=True)
```

```python
df[df.duplicated()]
```

```python
# Find the unique values in the state column
df['state'].unique()
```

```python
# Find the number of unique values in the state column
df['state'].nunique()
```

```python
# Find the NaN values in the dataframe
df.isnull().sum()
```

```python
# Find the number of NaN values in the City column
df['city'].isnull().sum()
```

```python
# Drop the NaN values in the City column
df.dropna(subset=['city'], inplace=True)
```

```python
# Value counts of the state column
df['state'].value_counts()
```

```python
# Value counts of the state column in percentage
df['state'].value_counts(normalize=True) * 100
```

```python
# Group by the state column and find the mean of the lat column
df.groupby('state')['latitude'].mean()
```

```python
# Value counts of the state column where the count is greater than 50
df['state'].value_counts()[df['state'].value_counts() > 50]
```

```python
# Find unique values of the city, state and country columns
df[['city', 'state', 'country']].drop_duplicates()
```

```python
# create a dataframe with dtype == object
df.select_dtypes(include='object')
```

```python
# use a for loop to print the unique values in the object columns
for col in df.select_dtypes(include='object').columns:
    print(df[col].unique())
```
