# Using Pandas Analysing Data

### Import Required Libraries
Import the necessary libraries, including pandas and matplotlib.

```python
# Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Enable inline plotting for Jupyter notebooks
%matplotlib inline
```

# Load the Dataset
Load the dataset from './data/seattle-temps.csv' using pandas.

```python
# Load the Dataset
df = pd.read_csv('./data/seattle-temps.csv', header=None, names=['Date', 'Temperature'])

# Display the first few rows of the dataframe to ensure it loaded correctly
df.head()
```

# Display Basic Information
Display the first few rows, data types, and basic information about the dataset.

```python
# Display Basic Information

# Display the last few rows of the dataframe
df.tail()

# Display the data types of each column
df.dtypes

# Display basic information about the dataframe
df.info()

# Display summary statistics for the dataframe
df.describe()
```

# Handle Missing Values
Check for and handle any missing values in the dataset.

```python
# Handle Missing Values

# Convert 'Temperature' column to numeric type
df['Temperature'] = pd.to_numeric(df['Temperature'], errors='coerce')

# Check for missing values in the dataset
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Handle missing values by filling them with the mean temperature
df['Temperature'].fillna(df['Temperature'].mean(), inplace=True)

# Verify that there are no more missing values
missing_values_after = df.isnull().sum()
print("Missing values in each column after handling:\n", missing_values_after)
```

# Generate Descriptive Statistics
Generate descriptive statistics for the dataset using pandas.

```python
# Generate Descriptive Statistics

# Generate descriptive statistics for the dataset
descriptive_stats = df.describe()

# Display the descriptive statistics
descriptive_stats
```

### Visualize Data with Scatter Plots
Create scatter plots to visualize the data using matplotlib.

```python
# Visualize Data with Scatter Plots

# Convert the 'Date' column to datetime format, skipping the first row
df = df.iloc[1:]
df['Date'] = pd.to_datetime(df['Date'])

# Create a scatter plot of temperature over time
plt.figure(figsize=(10, 6))
plt.scatter(df['Date'], df['Temperature'], alpha=0.5)
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (F)')
plt.grid(True)
plt.show()
```

# Visualize Data with Histograms
Create histograms to visualize the data distribution using matplotlib.

```python
# Visualize Data with Histograms

# Create a histogram to visualize the distribution of temperatures
plt.figure(figsize=(10, 6))
plt.hist(df['Temperature'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Distribution of Temperatures')
plt.xlabel('Temperature (F)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
